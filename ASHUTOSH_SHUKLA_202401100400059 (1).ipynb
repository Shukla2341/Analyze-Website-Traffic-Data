```python
# Import necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "website_traffic_50.csv"  # Upload the file in Colab
df = pd.read_csv('/content/website_traffic_large.txt')
```

```python
# Display first few rows
print("Dataset Overview:")
display(df.head())
```

```python
# -----------------------------------------------
# 1. Basic Traffic Analysis
# -----------------------------------------------
print("\nBasic Traffic Metrics:")
print(df[['Pageviews', 'Unique Visitors', 'Sessions']].describe())

```

```python
# Plot traffic trends over time
plt.figure(figsize=(12, 5))
sns.lineplot(x=df["Date"], y=df["Pageviews"], label="Pageviews")
sns.lineplot(x=df["Date"], y=df["Sessions"], label="Sessions")
plt.xticks(rotation=45)
plt.title("Website Traffic Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
# -----------------------------------------------
# 2. Bounce Rate & Session Analysis
# -----------------------------------------------
print("\nBounce Rate Analysis:")
print(df[['Bounce Rate (%)', 'Avg. Session Duration (s)']].describe())

plt.figure(figsize=(10, 5))
sns.histplot(df['Bounce Rate (%)'], bins=20, kde=True, color="red")
plt.title("Bounce Rate Distribution")
plt.xlabel("Bounce Rate (%)")
plt.show()
```

```python
# 4. Device-Based Traffic Insights
# -----------------------------------------------
device_traffic = df["Device Type"].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=device_traffic.index, y=device_traffic.values, palette="viridis")
plt.title("Traffic Distribution by Device Type")
plt.xlabel("Device Type")
plt.ylabel("Traffic Count")
plt.show()
```
